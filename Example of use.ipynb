```python
import numpy as np
import matplotlib.pyplot as plt
import STIT_functions as STIT
```

```python
P = STIT.Polygon(np.array([[0, 1, 1, 0, 0], [0, 0, 1, 1, 0]]).T) # Creates a polygon with n sides from an (n+1)x2 numpy array whose first and last row are identical.
```

```python
List_Polygon = STIT.STIT(P, 10) # Simulate the STIT tesselation inside the polygon P with final time T=10 and uniform distribution for the angles. Returns the list of polygons in the tesselation.
```

```python
# Drawing the tesselation

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
for Poly in List_Polygon:
    ax.plot(Poly.Points[:, 0], Poly.Points[:, 1], color="black");
xlim, ylim = P.Enclosing_Rectangle()
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.axis("off")
ax.set_aspect("equal");
```

```python
# Simulation inside a regular heptagon with a uniform distribution for angles

P = STIT.nGone(7)

List_Polygon = STIT.STIT(P, 5)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
for Poly in List_Polygon:
    ax.plot(Poly.Points[:, 0], Poly.Points[:, 1], color="black");
xlim, ylim = P.Enclosing_Rectangle()
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.axis("off")
ax.set_aspect("equal");
```

```python
# Simulation inside a regular heptagon with a distribution for angles supported on {0, pi/2} with equal probabilities.

P = STIT.nGone(7)

List_Polygon = STIT.STIT(P, 5, Dist="Side")

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
for Poly in List_Polygon:
    ax.plot(Poly.Points[:, 0], Poly.Points[:, 1], color="black");
xlim, ylim = P.Enclosing_Rectangle()
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.axis("off")
ax.set_aspect("equal");
```

```python
# Simulation inside a regular heptagon with a Von Mises distribution for angles taken modulo pi.

P = STIT.nGone(7)

List_Polygon = STIT.STIT(P, 5, Dist="VM", mu=np.pi/6, kappa=5)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
for Poly in List_Polygon:
    ax.plot(Poly.Points[:, 0], Poly.Points[:, 1], color="black");
xlim, ylim = P.Enclosing_Rectangle()
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.axis("off")
ax.set_aspect("equal");
```

```python
# Simulation inside a regular heptagon with a custom distribution for angles.

P = STIT.nGone(7)

def Random_angle():
    if np.random.rand() < 0.3:
        return np.pi/2
    return np.random.uniform(0, np.pi)

List_Polygon = STIT.STIT(P, 5, Dist="Custom", generator=Random_angle)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
for Poly in List_Polygon:
    ax.plot(Poly.Points[:, 0], Poly.Points[:, 1], color="black");
xlim, ylim = P.Enclosing_Rectangle()
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.axis("off")
ax.set_aspect("equal");
```
